```python
for sector_name, file_name in file_list.items():
    print(sector_name, file_name)
```

```python
period = 1.4200242
transit_time = 2456234.103218
transit_window = 1.66104* 1.5 /24. 
```

```python
plt.figure(figsize=(8,6))

for sector_name, file_name in file_list.items():

    lchdu = fits.open(file_name)
    sap_flux = lchdu[1].data['SAP_FLUX']
    sap_flux_error = lchdu[1].data['SAP_FLUX_ERR']
    pdcsap_flux = lchdu[1].data['PDCSAP_FLUX']
    pdcsap_flux_error = lchdu[1].data['PDCSAP_FLUX_ERR']
    quality_bitmask = lchdu[1].data['QUALITY']
    #print('BJDREFI header value:  {1:d}, comment: {0:s}'.format(lchdu[1].header.comments['BJDREFI'], lchdu[1].header['BJDREFI']))
    #print('BJDREFF header value: {1:f}, comment: {0:s}'.format(lchdu[1].header.comments['BJDREFF'], lchdu[1].header['BJDREFF']))
    time_array = lchdu[1].data['TIME'] + lchdu[1].header['BJDREFI'] +  lchdu[1].header['BJDREFF']

    finite_selection = np.isfinite(pdcsap_flux)

    temporal_exclusion = ((time_array > 2458774.4) & (time_array < 2458774.8)) | \
        ((time_array > 2458870.5) & (time_array < 2458871.5)) | \
        ((time_array > 2458883.5) & (time_array < 2458884.5)) | \
        ((time_array > 2459610.7) & (time_array < 2459611.2)) | \
        ((time_array > 2459704.5) & (time_array < 2459705.2)) | \
        ((time_array > 2459717.2) & (time_array < 2459717.8)) | \
        ((time_array > 2459825.0) & (time_array < 2459825.5)) | \
        ((time_array > 2459851.0) & (time_array < 2459856.8)) | \
        ((time_array > 2459866.0) & (time_array < 2459867.4)) | \
        ((time_array > 2459923.0) & (time_array < 2459923.5)) | \
        ((time_array > 2460370.5) & (time_array < 2460370.9)) | \
        ((time_array > 2460374.5) & (time_array < 2460375.1)) | \
        ((time_array > 2460394.6) & (time_array < 2460395.1)) | \
        ((time_array > 2460545.0) & (time_array < 2460545.6)) | \
        ((time_array > 2460552.0) & (time_array < 2460552.6)) | \
        ((time_array > 2460557.4) & (time_array < 2460559.8)) | \
        ((time_array > 2460584.4) & (time_array < 2460585.8)) | \
        ((time_array > 2460635) & (time_array < 2460638))


    conservative_selection =  ~(quality_bitmask > 0) & finite_selection & ~temporal_exclusion


    time = time_array[conservative_selection]
    flux = pdcsap_flux[conservative_selection]
    ferr = pdcsap_flux_error[conservative_selection]


    mask = transit_mask(
        time=time,
        period = period,
        duration = transit_window,
        T0=transit_time 
    )

    pdcsap_biweight_flatten_flux, pdcsap_biweight_flatten_model = flatten(
        time,
        flux,
        method='biweight',
        window_length=1.0, 
        break_tolerance=0.5, 
        return_trend = True,
        mask = mask,
    )

    pdcsap_huber_flatten_flux, pdcsap_huber_flatten_model = flatten(
        time,
        flux,
        method='huber',
        window_length=0.5, 
        break_tolerance=0.5, 
        return_trend = True,
        mask = mask,
    )

    pdcsap_average = (pdcsap_biweight_flatten_flux + pdcsap_huber_flatten_flux) / 2.
    pdcsap_average_model = (pdcsap_biweight_flatten_model + pdcsap_huber_flatten_model) / 2.
    error_average = ferr / pdcsap_average_model

    fileout = open(f'{sector_name}_pdcsap_processed.dat', 'w')
    fileout.write('# time flux error\n')
    for t, f, fe in zip(time, pdcsap_average, error_average):
        fileout.write('{0:12.6f} {1:18.6f} {2:18.6f}\n'.format(t, f, fe))
    fileout.close()

    plt.scatter(time_array[conservative_selection], pdcsap_flux[conservative_selection], c='C0',
                s=5, zorder=1)
    plt.scatter(time_array, pdcsap_flux, s=5, zorder=0, c='C1')
    plt.plot(time, pdcsap_average_model, color='C2')

plt.xlabel('BJD_TDB [d]')
plt.ylabel('e-/s')
plt.show()
```
